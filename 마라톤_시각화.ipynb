```python
import pandas as pd

marathon_2015_2017 = pd.read_csv("./datasets/marathon_results/marathon_2015_2017.csv")
```

```python
USA_runner = marathon_2015_2017[marathon_2015_2017.Country =='USA']
```

```python
plt.figure(figsize=(30,20))
```

```python
plt.figure(figsize=(40,20))

runner_state = sns.countplot('State',data=USA_runner)
runner_state.set_title('Number of Runner by State - USA', fontsize=18)
runner_state.set_xlabel('State', fontdict = {'size':16})
runner_state.set_ylabel('Number of Runner', fontdict = {'size':16})
plt.show()
```

```python
# 추세 

plt.figure(figsize=(40,20))

runner_state = sns.countplot('State',data=USA_runner, hue='Year')
runner_state.set_title('Number of Runner by State,Gender - USA', fontsize=18)
runner_state.set_xlabel('State', fontdict = {'size':16})
runner_state.set_ylabel('Number of Runner', fontdict = {'size':16})
plt.show()
```

# Dual axis Chart 만들기

```python
# 파레토 차트 matplotlib만사용한다
#  20세부터 60세까지 

#  18 to 60 까지 넣는다 
runner_1860 = marathon_2015_2017[marathon_2015_2017.Age.isin(range(18,60))]

# 연령별 카운팅 
runner_1860_counting = runner_1860['Age'].value_counts()
```

```python
runner_1860.info()
```

```python
# 연령별 인원수  index, value 형식 

runner_1860_counting
```

```python
runner_age = pd.DataFrame({
        'Age' : runner_1860_counting.index,
        'Count' : runner_1860_counting
})

print(runner_age)
```

```python

#  x축 정의 index 값을 보여줌 
x= runner_1860_counting.index


# 이걸 안하면 18이 맨 뒤에 가 string 으로 바꿔줘 
x = [str(i) for i in x]

# y 축 정의 
y = runner_1860_counting.values

# ratio dhk accumulated ratio 계산 
ratio = y / y.sum()
# 누적값 
ratio_sum = ratio.cumsum()


# figure size 정의 
# plt.figure(figsize=(40,20))
fig, barChart = plt.subplots(figsize=(20,10))

# barCahrt 생성 
barChart.bar(x,y)

# line barChart 생성 
# barchart 공유 twinx 
lineChart = barChart.twinx()
lineChart.plot(x, ratio_sum, '-ro' , alpha=0.5)

# 오른쪽 label 생성 
ranges = lineChart.get_yticks()
lineChart.set_yticklabels(['{:,.1%}'.format(x) for x in ranges])

#  라인차트 의 annotations 생성
ratio_sum_percentages = ['{0:.0%}'.format(x) for x in ratio_sum]
for i , txt in enumerate(ratio_sum_percentages):
    lineChart.annotate(txt, (x[i], ratio_sum[i]), fontsize=14)
    
#  labes 와 title 정의
barChart.set_xlabel('Age', fontdict= {'size':16})
barChart.set_ylabel('Number of runner', fontdict={'size':16})
plt.title('Pareto Chart - Number of runner by Age', fontsize=18)

plt.show()
```

```python
# 차트 정렬 

runner_age_sort = runner_age.sort_values(by=['Age'])
print(runner_age_sort)
```

```python
# y 축 정의 
y = runner_age_sort['Count'] 
# ratio dhk accumulated ratio 계산 
ratio = y / y.sum()
# 누적값 
ratio_sum = ratio.cumsum()

# x 값은 배열인데, y 배열이 아니어서 에러가 난다 그래서 y값을 재정의 
y_ratio = [i for i in ratio_sum]

print(ratio_sum)
print(y_ratio)

# figure size 정의 
# plt.figure(figsize=(40,20))
fig, barChart = plt.subplots(figsize=(20,10))

# barCahrt 생성 
barChart.bar(x,y)

# line barChart 생성 
# barchart 공유 twinx 
lineChart = barChart.twinx()
lineChart.plot(x, ratio_sum, '-ro' , alpha=0.5)

# 오른쪽 label 생성 
ranges = lineChart.get_yticks()
lineChart.set_yticklabels(['{:,.1%}'.format(x) for x in ranges])

#  라인차트 의 annotations 생성
ratio_sum_percentages = ['{0:.0%}'.format(x) for x in ratio_sum]
for i , txt in enumerate(ratio_sum_percentages):
    lineChart.annotate(txt, (x[i], ratio_sum[i]), fontsize=14)
    
#  labes 와 title 정의
barChart.set_xlabel('Age', fontdict= {'size':16})
barChart.set_ylabel('Number of runner', fontdict={'size':16})
plt.title('Pareto Chart - Number of runner by Age', fontsize=18)

plt.show()
```
